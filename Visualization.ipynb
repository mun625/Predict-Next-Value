# Visualizae some integer sequences
At first, I will plot some sequences into 2-dimensional coordinates so that can get insights.

```python
# Import library and define some function that need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def ListToGraph(intList):
    if type(intList) != list:
        raise ValueError("A sequence's type should be list")
    if not all(isinstance(n, int) or isinstance(n, long) for n in intList):
        raise ValueError("The element of sequence should be integer type")
    #intList = intList[0:-1]
    plt.plot(intList, 'ro')
    plt.plot(intList)
    plt.ylabel('value')
    plt.show()
```

```python
# Read the sequence data
train_data = pd.read_csv('data/train.csv')

# Set indexes that we want to visualize as graphs
start_index = 150
last_index = 155

for index in range(start_index, last_index+1):
    my_list = train_data.iloc[index]['Sequence'].split(',')
    my_list = map(int,my_list)
    ListToGraph(my_list)
```
